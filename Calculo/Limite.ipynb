# Introdução ao Calculo Diferencial:

    A base da otimização de redes neurais é descida de gradiente,
    que por sua vez depende da capacidade de encontrar derivadas
    automaticamente. Em primeiro lugar, é útil entender quão simples
    tomar uma derivada pode ser, do ponto de vista computacional.

**Derivada Unidimensional**

    Se lembramos da definição de Derivada, temos que:

image.png

    Onde h é um valor infinitesimal, cuja a teoria do limite
    define como um valor que tende a zero.

```python
# Função:
def f1(x):
  return 3*(x**2)-4*x+5

# Valor infinitesimal:
h = 0.0000001

# Ponto da Função:
x = 3

# derivada:
(f1(x+h)-f1(x))/h
```

**Derivada Parcial**

    Para a derivada parcial, a ideia é a mesma, vamos trabalhar com essa mesma
    ideia, porém com pequenas alterações.

```python
# Função:
def f2(a,b,c,d):
  return (a*b+c)*d

# Infinitesimal:
h = 0.0000001

# ponto:
a = 2
b = -3
c = 10
d = -2

# Derivadas:
print(f" Derivada em a: {(f2(a+h, b, c, d)-f2(a, b, c, d))/h}")
print(f" Derivada em b: {(f2(a, b+h, c, d)-f2(a, b, c, d))/h}")
print(f" Derivada em c: {(f2(a, b, c+h, d)-f2(a, b, c, d))/h}")
print(f" Derivada em d: {(f2(a, b, c, d+h)-f2(a, b, c, d))/h}")
```

**Conclusão**

    A principio esse pode ser o primeiro jeito que você pense em resolver
    esse tipo de situação. É uma solução válido, porem, note como a derivada
    não é exata, há um erro associado... E isso pode ser um problema...
    Um grande problema em alguns casos.

    Como já dito, a ideia de se estudar a contrução matemática de gradientes
    no computador veio com a necessidade de utiliza-los como otimizador
    de um rede neural, porém, esse jeito de calcular não é bom, porque
    ele carrega essa incerteza para o modelo. Imagine fazer um back propagation
    com essa forma de se calcular o gradiente, não seria interessante.
